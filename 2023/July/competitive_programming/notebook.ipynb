```python
# Infinite Loop using continue

i = 0
while i<5:
    if i==2:
        continue
    else:
        print(i,end=",")
        i+=1

# 1. 1,2,3,4
# 2. 1,2,4
# 3. Time Limit Error
# 4. 1

# In the above code snippet the value of i will be equal to 2 then we are performing a continue but not incrementing the value of i will be equal to 2 again and again which results in a Runtime Error.
```

```python
# Infinite Loop

i = 0
while(i<5):
    print(i,end=" ")

# In the above code snippet we're not increasing the value of i so the loop will run infinite times.
```
